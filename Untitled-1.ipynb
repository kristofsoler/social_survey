### The surveys for each column each have inconsitant scales, some are from 0-4 some 1-10. The non answers (like did not respond etc) are coded as different numbers  sometimes 66,77,88 or 6,7,8. To detect the scales in each column, the minimum number in the column was detected and then a search for the next whole number. When the next consective number was not detected that was considered the end of the scale. All numbers outside this scale were considered non answers oand transfered til nan. The scale ranges were checked against the dataset documentation. 

Get the scale ranges

```python
def identify_scale(column):
    # Drop missing values and get unique sorted values
    sorted_values = sorted(column.dropna().unique())
    
    # Check if the column has any valid values
    if not sorted_values:
        return None, None
    
    # Initialize variables to track the start and end of the scale
    scale_start = sorted_values[0]
    scale_end = sorted_values[0]
    
    # initialize variables to track the previous value in the sequence
    prev_value = sorted_values[0]
    
    # Search for consecutive higher numbers
    for value in sorted_values[1:]:
        if value == prev_value + 1:
            scale_end = value
        else:
            # break if a gap is found
            break
        prev_value = value
    
    return scale_start, scale_end
```

```python

scale_ranges = {}

# Process each column to identify the scale
for column in data1.columns:
    scale_start, scale_end = identify_scale(data1[column])
    scale_ranges[column] = (scale_start, scale_end)

# Display the identified scale ranges for each column
for column, (scale_start, scale_end) in scale_ranges.items():
    print(f"Column '{column}': Scale range = {scale_start}-{scale_end}")
```

Convert values outide the scale range for each column to nan

```python
def convert_outside_scale_to_nan(column, scale_start, scale_end):
    # Replace values outside the scale range with NaN
    column[(column < scale_start) | (column > scale_end)] = np.nan
    return column

# Iterate through each column in the DataFrame
for column in data1.columns:
    # Get the scale range for the current column
    scale_start, scale_end = scale_ranges[column]
    
    # Convert values outside the scale range to NaN
    data1[column] = convert_outside_scale_to_nan(data1[column], scale_start, scale_end)
```

### Next step will be too deal wit the missing data. In this case we are performing  bootstrapping and imputation for each column in the DataFrame data1. Bootstrap samples are generated from non-missing values, and missing values are imputed with the mean of the bootstrap sample. the mean is calculated across the bootstrapped imputed values for each column.

```python
import pandas as pd
import numpy as np

# mean imputation method
def mean_imputation(column):
    mean_value = np.nanmean(column)  #  mean excluding NaN values
    return column.fillna(mean_value)

# Perform bootstrapping and imputation for each column
bootstrap_iterations = 1000
imputed_data = data1.copy()  # Create a copy of the original data for imputation

for column in data1.columns:
    non_missing_values = data1[column].dropna()
    if non_missing_values.empty:
        # If the column has no non-missing values, skip bootstrapping
        continue
    
    for _ in range(bootstrap_iterations):
        bootstrap_sample = np.random.choice(non_missing_values, size=len(data1[column]), replace=True)
        bootstrap_mean = np.mean(bootstrap_sample)
        imputed_data[column].fillna(bootstrap_mean, inplace=True)

# calculate mean across bootstrapped imputed 
final_imputed_data = imputed_data.mean()  # Use mean as the aggr imputed value

# Replace missing values in the original dataset with the aggregated values 
data2 = data1.fillna(final_imputed_data, inplace=False)


print(data2)
```
